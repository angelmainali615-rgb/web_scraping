# Introduction

- Web scraping is the automated collection of data from websites.
- It is widely used in data science, research, journalism, and industry.
- Scraping helps extract publicly available data that is not in structured formats.
- This notebook covers requesting web pages, parsing HTML, and extracting data.
- Ethical and responsible scraping practices are essential in real-world use.

# Ethics of Web Scraping

Although web scraping often involves publicly accessible data, it raises important ethical and legal considerations. Responsible scraping requires respecting both website owners and users.

Key ethical principles include:

- **Respect website policies**: Always review a website’s `robots.txt` file and terms of service to understand what is permitted.
- **Avoid excessive requests**: Sending too many requests in a short period can overload servers. Implement rate limiting and delays when scraping.
- **Do not scrape sensitive data**: Personal, private, or confidential information should never be collected without explicit permission.
- **Attribute data sources**: When using scraped data for research or publication, properly credit the original source.
- **Use data responsibly**: Scraped data should be used in ways that do not harm individuals, organizations, or communities.

Ethical web scraping balances technical capability with responsibility, ensuring that data collection practices are fair, transparent, and respectful.

```python
import requests
from bs4 import BeautifulSoup

# The URL of a simple website to scrape
url = "http://books.toscrape.com/";

# 1. Fetch the web page content
try:
    response = requests.get(url)
    response.raise_for_status()

    # 2. Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
   
    # 3. Extract the data
    # We want to get the title of every book on the page.
    # By inspecting the website, we find that book titles are in <h3> tags
    # which are inside <article class="product_pod"> tags.

   
    book_titles = []
    book_prices = []
    # Find all 'article' tags with the class 'product_pod'
    for book in soup.find_all('article', class_='product_pod'):
        # Inside each article, find the 'h3' tag, then the 'a' tag, and get its title attribute
        title = book.h3.a['title']
        href = book.h3.a['href']
        new_title = title + '  ' + href
        book_titles.append(new_title)

       
        price = book.find('p', class_="price_color").text
        book_prices.append(price)
       
       
    print("--- Found Book Titles ---")
    for i, (title, price) in enumerate(zip(book_titles,book_prices),start=1): # Print first 5
        print(f"{i}. {title}, Price:{price}")

except requests.exceptions.RequestException as e:
    print(f"Error fetching URL: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv 
import time 
# Sitemap URL (contains all public allowed pages)
URL="https://realpython.com/"

# Add headers to behave like a real browser
headers={
    "User_Agent": "Mywebscraper/1.0 (Educational;contact:angelmainali615@gmail.com)" 
}
# Send request
response=requests.get(URL,headers=headers)
# Check if request succeeded
if response.status_code != 200:
    raise Exception("Failed to load page")

#converts raw HTML text into a searchable Python object so you can find tags like <div>, <h1>, <a>, etc.
soup=BeautifulSoup(response.text,"html.parser")

# Each tutorial is inside a card
cards=soup.find_all("div", class_="card")

results=[]

for card in cards:
    title_tags=card.find("h2")
    title=title_tags.text.strip() if title_tags else "N/A"

    link_tags=card.find("a")
    link=(
        "http://realpython.com/" + link_tags.get('href')
        if link_tags else "N/A"
    )

    desc_tags=card.find("p")
    description=desc_tags.text.strip() if desc_tags else "N/A"

    results.append([title,link,description])

# Save the scraped results into a CSV file
# "with" ensures the file is properly closed after writing
with open("realpython_allowed_tutorials.csv", "w", newline="", encoding="utf-8") as f:
    # Create a CSV writer object that knows how to format rows
    writer = csv.writer(f)
    
    # Write the header row (column names) to the CSV file
    writer.writerow(["Title", "URL", "Description"])
    
    # Write all scraped data rows at once
    # Each item in 'results' represents one row in the CSV
    writer.writerows(results)

```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time

# Sitemap URL (contains all public allowed pages)
SITEMAP_URL = "https://realpython.com/sitemap.xml"

headers = {
    "User-Agent": "MyWebScraper/1.0 (Educational Purpose)"
}

# Fetch sitemap.xml
response = requests.get(SITEMAP_URL, headers=headers)
if response.status_code != 200:
    raise Exception("Failed to fetch sitemap.xml")

# Parse XML
soup = BeautifulSoup(response.content, "xml")  # note: xml parser

# Find all <loc> tags (these contain URLs)
urls = [loc.text for loc in soup.find_all("loc")]

# Filter only tutorials (allowed content)
tutorial_urls = [url for url in urls if "/tutorials/" in url]

# Prepare results
results = []

for url in tutorial_urls:
    # Simple title extraction from URL (optional)
    title = url.split("/")[-2].replace("-", " ").title()
    results.append([title, url])

    time.sleep(0.05)  # polite delay

# Save to CSV
with open("realpython_tutorials_sitemap.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "URL"])
    writer.writerows(results)

print(f"Scraped {len(results)} tutorials successfully from sitemap!")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time


# Fetch sitemap.xml
SITEMAP_URL = "https://realpython.com/sitemap.xml"

headers = {
    "User-Agent": "MyWebScraper/1.0 (Educational Purpose)"
}

response = requests.get(SITEMAP_URL, headers=headers)
if response.status_code != 200:
    raise Exception("Failed to fetch sitemap.xml")


#  Parse XML sitemap
soup = BeautifulSoup(response.content, "xml")  # Parse XML (not HTML)
all_urls = [loc.text for loc in soup.find_all("loc")]  # Extract all <loc> URLs


# Filter only allowed tutorial URLs

tutorial_urls = [url for url in all_urls if "/tutorials/" in url]

# Loop through URLs to extract metadata
results = []

for url in tutorial_urls:
    
    
    # Fetch each tutorial page (allowed)
    page = requests.get(url, headers=headers)
    if page.status_code != 200:
        print(f"Failed to fetch {url}")
        continue
    
    page_soup = BeautifulSoup(page.text, "html.parser")
    
    
    # Extract Title-
    title_tag = page_soup.find("h1")
    title = title_tag.text.strip() if title_tag else url.split("/")[-2].replace("-", " ").title()
    
    # Extract Short Description
    desc_tag = page_soup.find("meta", attrs={"name": "description"})
    description = desc_tag.get("content").strip() if desc_tag else "N/A"
    
    
    # Extract Author (if available)
    author_tag = page_soup.find("span", class_="author-name")
    author = author_tag.text.strip() if author_tag else "N/A"
    
    
    # Extract Category/Tags (if available)
    tags = page_soup.find_all("a", class_="tag")
    categories = ", ".join([t.text.strip() for t in tags]) if tags else "N/A"
    
    
    # Extract Last Modified Date (if available)

    time_tag = page_soup.find("time")
    lastmod = time_tag.get("datetime") if time_tag else "N/A"
    
    # Append to results
    results.append([title, url, author, categories, lastmod, description])

#  Save results to CSV
with open("realpython_full_tutorials.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "URL", "Author", "Category/Tags", "Last Modified", "Description"])
    writer.writerows(results)

print(f"Scraped {len(results)} tutorials successfully!")
```
